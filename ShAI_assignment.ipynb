**1. Basic Data Exploration:**

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('/content/Salaries.csv')
df.head()
```

```python
# Display the number of rows and columns
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])

# Display data types of each column and check for missing values
print("\nData types and missing values:")
print(df.info())
```

```python
df.columns
```

**2. Descriptive Statistics:**

```python
# Calculate mean, median, mode, minimum, and maximum salary
mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()
mode_salary = df['TotalPay'].mode()[0]  # Mode returns a Series, so we take the first element
min_salary = df['TotalPay'].min()
max_salary = df['TotalPay'].max()

print("\nBasic Statistics for TotalPay:")
print("Mean Salary:", mean_salary)
print("Median Salary:", median_salary)
print("Mode Salary:", mode_salary)
print("Minimum Salary:", min_salary)
print("Maximum Salary:", max_salary)

# Calculate range
salary_range = df['TotalPay'].max() - df['TotalPay'].min()

# Calculate standard deviation
salary_std = df['TotalPay'].std()

print("\nSalary Range:", salary_range)
print("Standard Deviation of Salary:", salary_std)
```

**3. Data Cleaning:**

Handling missing data depends on the nature of the missing values. Common methods include filling with mean/median or dropping rows.

```python
# Replace missing values in 'BasePay' with the mean
df.fillna({'BasePay': df['BasePay'].mean()}, inplace=True)
print(df)
```

**4. Basic Data Visualization:**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of salaries
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPay'], bins=30, kde=True)
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()

# Pie chart of departments
department_counts = df['Agency'].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

**5. Grouped Analysis:**

```python
# Group by 'Year' and 'Agency' and calculate summary statistics
grouped_data = df.groupby(['Year', 'Agency'])['TotalPay'].agg(['mean', 'count', 'min', 'max'])

# Display the grouped data
print("\nSummary Statistics by Year and Agency:")
print(grouped_data)
```

**6. Simple Correlation Analysis:**

```python
# Check correlation between 'TotalPay' and 'Benefits'
correlation = df['TotalPay'].corr(df['Benefits'])
print("\nCorrelation between TotalPay and Benefits:", correlation)

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Benefits', y='TotalPay', data=df)
plt.title('Scatter Plot of TotalPay vs Benefits')
plt.xlabel('Benefits')
plt.ylabel('Total Pay')
plt.show()
```
